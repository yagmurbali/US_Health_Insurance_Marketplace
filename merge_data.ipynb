```python
planAttributes= plan_attributes[['IssuerId','MetalLevel', 'DentalOnlyPlan', 'MarketCoverage', 'StateCode']]
planAttributes.head()
```

```python
rate_= rate[['Age', 'Tobacco', 'IndividualRate', 'IndividualTobaccoRate']]
rate_.head()
```

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.truncate.html
rate_ = rate_.truncate(after=14999)
```

```python
benefitCost= benefit_cost[['IssuerId','BenefitName', 'IsCovered', 'IsEHB', 'QuantLimitOnSvc', 'IsExclFromInnMOOP', 'IsExclFromOonMOOP']]
benefitCost.head()
```

```python
benefitCost = benefitCost.truncate(after=14999)
```

```python
merged = pd.concat([rate_, benefitCost], axis=1)
```

```python
healthInsurance = pd.merge(merged, planAttributes, on='IssuerId')
```
